```python
import random
```

```python
def generate_argos_maze(width=10, height=10, arena_size=10.0, wall_thickness=0.1):
    grid = [[{'N': True, 'W': True, 'visited': False} for _ in range(width)] for _ in range(height)]
    
    def carve_passages(cx, cy):
        grid[cy][cx]['visited'] = True
        directions = [('N', 0, -1, 'S'), ('S', 0, 1, 'N'), ('E', 1, 0, 'W'), ('W', -1, 0, 'E')]
        random.shuffle(directions)
        
        for direction, dx, dy, opposite in directions:
            nx, ny = cx + dx, cy + dy
            if 0 <= nx < width and 0 <= ny < height and not grid[ny][nx]['visited']:
                if direction == 'N': grid[cy][cx]['N'] = False
                elif direction == 'W': grid[cy][cx]['W'] = False
                elif direction == 'S': grid[ny][nx]['N'] = False
                elif direction == 'E': grid[ny][nx]['W'] = False
                carve_passages(nx, ny)

    carve_passages(0, 0)
    
    cell_w = arena_size / width
    cell_h = arena_size / height
    xml_output = []
    obs_idx = 0

    for y in range(height):
        for x in range(width):
            cx = -arena_size / 2 + (x * cell_w) + (cell_w / 2)
            cy = arena_size / 2 - (y * cell_h) - (cell_h / 2)

            if grid[y][x]['N'] and y != 0:
                xml_output.append(f'    <box id="maze_n_{obs_idx}" size="{cell_w + wall_thickness},{wall_thickness},0.5" movable="false">\n      <body position="{cx},{cy + cell_h/2},0" orientation="0,0,0"/>\n    </box>')
                obs_idx += 1
                
            if grid[y][x]['W'] and x != 0:
                xml_output.append(f'    <box id="maze_w_{obs_idx}" size="{wall_thickness},{cell_h + wall_thickness},0.5" movable="false">\n      <body position="{cx - cell_w/2},{cy},0" orientation="0,0,0"/>\n    </box>')
                obs_idx += 1

    start_x = -arena_size / 2 + (0 * cell_w) + (cell_w / 2)
    start_y = arena_size / 2 - (0 * cell_h) - (cell_h / 2)
    start = (start_x, start_y)

    end_x = -arena_size / 2 + ((width - 1) * cell_w) + (cell_w / 2)
    end_y = arena_size / 2 - ((height - 1) * cell_h) - (cell_h / 2)
    end = (end_x, end_y)

    return "\n".join(xml_output), start, end
```

```python
maze, start, end = generate_argos_maze(10, 10, 10.0)
print(maze)
print(start)
print(end)
```
